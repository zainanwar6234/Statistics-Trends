```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Load data from Dataset
df = pd.read_csv('./dataset/T20WorldCup2024Stats.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
# Basic Stats
print("Basic Statistics:\n", df.describe())
```

```python
# Select only numeric data for the correlation matrix
numericData = df.select_dtypes(include=[np.number])
```

```python
print("\nCorrelation Matrix:")
numericData.corr()
```

```python
# Plotting histogram of runs
plt.figure(figsize=(8, 6))
plt.hist(numericData['runs'], bins=range(1,120,20), color='dodgerblue', edgecolor='black')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.title('Distribution of Runs Scored')
plt.show()
```

```python
# Scatter plot of balls vs runs
plt.figure(figsize=(8, 6))
plt.scatter(numericData['balls'], numericData['runs'], color='green', alpha=0.6)
plt.xlabel('Balls Faced')
plt.ylabel('Runs Scored')
plt.title('Relationship between Balls Faced and Runs Scored')
plt.show()
```

```python
# Compute correlation matrix
import seaborn as sb
corr_matrix = numericData[['battingPos', 'runs', 'balls', '4s', '6s']].corr()
plt.figure(figsize=(8, 6))
sb.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
from scipy.stats import skew, kurtosis
import seaborn as sb

for column in numericData:
    plt.figure(dpi=120)
    sb.histplot(numericData[column], kde=True, color='skyblue', edgecolor='black')

    
    col_skewness = skew(numericData[column])
    
    # Add mean and median lines
    plt.axvline(numericData[column].mean(), color='red', linestyle='--', label=f'Mean: {numericData[column].mean():.2f}')
    plt.axvline(numericData[column].median(), color='green', linestyle='-', label=f'Median: {numericData[column].median():.2f}')
    
    # Add title and labels
    plt.title(f'Distribution of {column} (Skewness: {col_skewness:.2f})')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```
